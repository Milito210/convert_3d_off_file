Evaluate the model

To evaluate our model on the test set, we compute the most likely class for each example using the softmax outputs

Because the classes are inbalanced, the standard accuracy metric for ModelNet10 is average per-class accuracy. We can compute this from the confusion matrix:

```python
!python scripts/evaluate.py
```

```python
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, cohen_kappa_score
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
# ng dn thay i ty theo mi trng local hoc Google Colab
model_path = '/content/drive/MyDrive/convert_3d_off_file/models/model.h5'
data_path = '/content/drive/MyDrive/convert_3d_off_file/data/modelnet10.npz'

# Load d liu
print("Loading data...")
data = np.load(data_path)
X_test, y_test = data['X_test'], data['y_test']

# Load model
from keras.models import load_model
model = load_model(model_path)

# D on
print("Predicting...")
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)

# Tnh cc thng s nh gi
print("Evaluating...")
accuracy = accuracy_score(y_test, y_pred_classes)
kappa = cohen_kappa_score(y_test, y_pred_classes)

# Bo co chi tit precision, recall, F1-score
report = classification_report(y_test, y_pred_classes, target_names=[
    'bathtub', 'bed', 'chair', 'desk', 'dresser', 'monitor', 'night_stand', 'sofa', 'table', 'toilet'
])

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred_classes)

# Hin th kt qu
print("Accuracy:", accuracy)
print("Cohen's Kappa:", kappa)
print("Classification Report:\n", report)

# V Confusion Matrix
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=[
    'bathtub', 'bed', 'chair', 'desk', 'dresser', 'monitor', 'night_stand', 'sofa', 'table', 'toilet'
], yticklabels=[
    'bathtub', 'bed', 'chair', 'desk', 'dresser', 'monitor', 'night_stand', 'sofa', 'table', 'toilet'
])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```
